```python
from datetime import datetime, timedelta

registro_de_datos = {}
clave_p = 1
biblioteca_de_citas = {}
folio_cita = 0
biblioteca_de_citas_programadas = {}

def imprimir_menu():
    print("\nBienvenid@ a nuestro menu de opciones: \n",
          "1.- Registro de pacientes\n",
          "2.- Programacion de citas\n",
          "3.- Realizacion de citas programadas\n",
          "4.- Consultas y reportes\n",
          "5.- Salir del sistema\n")

def registro_pacientes(registro_de_datos):
    global clave_p
    primer_apellido = input("¿Cual es tu primer apellido?\n")
    while not primer_apellido or not primer_apellido.isalpha():
        print("El primer apellido es obligatorio y debe contener solo letras.")
        primer_apellido = input("Ingrese el primer apellido: ")

    segundo_apellido = input("¿Cual es tu segundo apellido? (puede omitirlo si no lo posee)\n")
    nombre = input("¿Cual es tu nombre?\n")
    while not nombre or not nombre.isalpha():
        print("El nombre es obligatorio y debe contener solo letras.")
        nombre = input("Ingrese el nombre: ")

    while True:
        try:
            fecha_str = input("Ingrese su fecha de nacimiento (mm/dd/aaaa): ")
            fecha_procesada = datetime.strptime(fecha_str, "%m/%d/%Y").date()
            fecha_actual = datetime.now().date()

            if fecha_procesada > fecha_actual:
                print("Error: La fecha ingresada debe ser anterior al día actual.")
                continue
            elif fecha_procesada.month > 12:
                print("Error: El mes ingresado es inválido.")
                continue
            else:
                print("Fecha ingresada correctamente")
                fecha_procesada = fecha_procesada.strftime('%m/%d/%Y')
                break
        except ValueError:
            print("Error: Formato de fecha invalido. Por favor, ingrese la fecha en el formato indicado.")

    registro_de_datos[clave_p] = [primer_apellido, segundo_apellido, nombre, fecha_procesada]
    print(registro_de_datos)
    clave_p += 1

def registro_citas(biblioteca_de_citas):
    global folio_cita
    folio_cita += 1

    while True:
        seleccion_clave = input("Digame la clave del paciente: ")
        try:
            seleccion_clave = int(seleccion_clave)
            if seleccion_clave in registro_de_datos:
                break
            else:
                print("Su clave no esta registrada")
                continue
        except ValueError:
            print("El dato que ingreso no es de tipo numerico")
            continue

    fecha_limite = datetime.today() + timedelta(days=60)
    fecha_limite_print = fecha_limite.strftime('%m/%d/%Y')
    print(f"Fecha limite de cita {fecha_limite_print}")

    while True:
        fecha_cita_str = input("Introduzca la fecha en la que quiera consultar: mm/dd/aaaa ")

        try:
            fecha_cita = datetime.strptime(fecha_cita_str, "%m/%d/%Y").date()

            if fecha_cita <= fecha_limite.date():
                fecha_limite = fecha_cita
                fecha_limite_print = fecha_limite.strftime('%m/%d/%Y')
                print("Su fecha es valida")
                break
            else:
                print("Ingrese una fecha no mayor a 60 dias despues del dia de hoy")
                continue
        except ValueError:
            print("La fecha no coincide con el formato indicado")
            continue

    while True:
        try:
            turno_de_cita = int(input("Introduzca el horario que desea: (1-Mañana 2-Tarde 3-Noche) "))
            if turno_de_cita in [1, 2, 3]:
                print("Su horario se registro correctamente \n")
                break
            else:
                print("Su opcion no coincide con ninguna de las opciones")
        except ValueError:
            print("El valor debe ser numerico")
            continue

    biblioteca_de_citas[folio_cita] = [seleccion_clave, fecha_limite, turno_de_cita]
    print(f"Cita programada con folio: {folio_cita}")

def cita():
    global folio_cita
    hora_llegada = datetime.now().strftime("%H:%M:%S")

    while True:
        seleccion_folio_cita = input("Ingrese el folio de la cita: ")
        try:
            seleccion_folio_cita = int(seleccion_folio_cita)
            if seleccion_folio_cita in biblioteca_de_citas:
                break
            else:
                print("Su folio no esta registrado")
                continue
        except ValueError:
            print("El folio debe ser un número entero.")

    while True:
        seleccion_clave = input("Digame la clave del paciente: ")
        try:
            seleccion_clave = int(seleccion_clave)
            if seleccion_clave in registro_de_datos:
                break
            else:
                print("Su clave no esta registrada")
                continue
        except ValueError:
            print("El dato que ingreso no es de tipo numerico")
            continue

    while True:
        peso = input("Ingrese el peso del paciente en kilogramos: ")
        try:
            peso = float(peso)
            if peso <= 0:
                print("El peso no puede ser menor o igual a cero, ingrese nuevamente")
                continue
            else:
                print("Peso registrado")
                break
        except ValueError:
            print("El dato ingresado no es de tipo numerico")
            continue

    while True:
        estatura = input("Ingrese la estatura en centimetros: ")
        try:
            estatura = float(estatura)
            if estatura <= 0:
                print("La estatura no puede ser menor o igual a cero, ingrese nuevamente")
                continue
            else:
                print("Estatura registrada")
                break
        except ValueError:
            print("El dato ingresado no es de tipo numerico")
            continue

    biblioteca_de_citas_programadas[seleccion_folio_cita] = {
        'clave_p': seleccion_clave,
        'hora_llegada': hora_llegada,
        'peso': peso,
        'estatura': estatura
    }
    print(biblioteca_de_citas_programadas)

def generar_reporte_citas_por_periodo():
    print("\nReporte de Citas por Período")
    print("---------------------------------")

    while True:
        try:
            fecha_inicio_str = input("Ingrese la fecha de inicio (mm/dd/aaaa): ")
            fecha_inicio = datetime.strptime(fecha_inicio_str, "%m/%d/%Y").date()
            break
        except ValueError:
            print("La fecha no coincide con el formato indicado. Inténtelo nuevamente.")

    while True:
        try:
            fecha_fin_str = input("Ingrese la fecha de fin (mm/dd/aaaa): ")
            fecha_fin = datetime.strptime(fecha_fin_str, "%m/%d/%Y").date()
            if fecha_fin < fecha_inicio:
                print("La fecha de fin debe ser posterior a la fecha de inicio.")
                continue
            break
        except ValueError:
            print("La fecha no coincide con el formato indicado. Inténtelo nuevamente.")

    print("\nCitas programadas en el período:")
    print("{:<10} {:<15} {:<20} {:<10}".format("Folio", "Clave Paciente", "Fecha de Cita", "Turno"))
    print("-----------------------------------------------")

    for folio, datos_cita in biblioteca_de_citas.items():
        fecha_cita = datos_cita[1]
        if fecha_inicio <= fecha_cita <= fecha_fin:
            print("{:<10} {:<15} {:<20} {:<10}".format(folio, datos_cita[0], fecha_cita.strftime('%m/%d/%Y'), datos_cita[2]))

def generar_reporte_pacientes():
    print("\nReporte de Pacientes Registrados")
    print("---------------------------------")
    print("{:<15} {:<15} {:<20} {:<15} {:<15}".format("Clave Paciente", "Primer Apellido", "Segundo Apellido", "Nombre", "Fecha de Nacimiento"))
    print("----------------------------------------------------------------------------------------")

    for clave, datos_paciente in registro_de_datos.items():
        print("{:<15} {:<15} {:<20} {:<15} {:<15}".format(clave, datos_paciente[0], datos_paciente[1], datos_paciente[2], datos_paciente[3]))

def listado_completo_pacientes():
    print("\nListado completo de pacientes:")
    print("Clave | Apellido Paterno | Apellido Materno | Nombre | Fecha de Nacimiento")
    for clave, paciente in registro_de_datos.items():
        print(f"{clave:<6} | {paciente[0]:<16} | {paciente[1]:<16} | {paciente[2]:<6} | {paciente[3]}")


def buscar_paciente_por_clave():
    while True:
        clave_buscada = input("Ingrese la clave del paciente que desea buscar: ")
        if clave_buscada.isnumeric():
            clave_buscada = int(clave_buscada)
            if clave_buscada in registro_de_datos:
                datos_paciente = registro_de_datos[clave_buscada]
                print("\nDatos del paciente con clave {}:".format(clave_buscada))
                print("{:<15} {:<15} {:<20} {:<15} {:<15}".format("Clave Paciente", "Primer Apellido", "Segundo Apellido", "Nombre", "Fecha de Nacimiento"))
                print("----------------------------------------------------------------------------------------")
                print("{:<15} {:<15} {:<20} {:<15} {:<15}".format(clave_buscada, datos_paciente[0], datos_paciente[1], datos_paciente[2], datos_paciente[3]))
                break
            else:
                print("\nNo se encontró un paciente con la clave {}".format(clave_buscada))
        else:
            print("\nIngrese un valor numérico válido.")

def buscar_pacientes_por_apellidos_y_nombres():
    while True:
        apellido_buscado = input("Ingrese el apellido del paciente que desea buscar: ")
        nombre_buscado = input("Ingrese el nombre del paciente que desea buscar: ")

        if apellido_buscado.isalpha() and nombre_buscado.isalpha():
            encontrados = []

            for clave, datos_paciente in registro_de_datos.items():
                primer_apellido = datos_paciente[0]
                segundo_apellido = datos_paciente[1]
                nombre = datos_paciente[2]

                if apellido_buscado.lower() in primer_apellido.lower() and nombre_buscado.lower() in nombre.lower():
                    encontrados.append((clave, datos_paciente))

            if encontrados:
                print("\nResultados de la búsqueda:")
                print("f{:<15} {:<15} {:<20} {:<15} {:<15}".format("Clave Paciente", "Primer Apellido", "Segundo Apellido", "Nombre", "Fecha de Nacimiento"))
                print("----------------------------------------------------------------------------------------")
                for clave, datos_paciente in encontrados:
                    print("f{:<15} {:<15} {:<20} {:<15} {:<15}".format(clave, datos_paciente[0], datos_paciente[1], datos_paciente[2], datos_paciente[3]))
                break
            else:
                print("\nNo se encontraron pacientes con los apellidos y nombres proporcionados.")
        else:
            print("\nIngrese valores alfabéticos válidos para el apellido y nombre.")

def confirmar_salida():
    while True:
        respuesta = input("¿Está seguro de que desea salir del sistema? (S/N): ").lower()
        if respuesta in ['s', 'n']:
            return respuesta == 's'
        else:
            print("\nIngrese una opción válida (S/N).")

while True:
    imprimir_menu()

    try:
        opcion_menu = int(input("Ingrese la opcion que desea realizar: "))
    except ValueError:
        print("Debera de ser un valor numerico y dentro del rango de opciones")
        continue

    if opcion_menu == 1:
        registro_pacientes(registro_de_datos)

    elif opcion_menu == 2:
        if clave_p == 0:
            print("Error: No hay pacientes registrados. Registre al menos un paciente antes de programar citas.")
            continue
        registro_citas(biblioteca_de_citas)

    elif opcion_menu == 3:
        if not biblioteca_de_citas:
            print("Error: No hay citas programadas. Programe al menos una cita antes de realizarla.")
            continue
        cita()

    elif opcion_menu == 4:
        if not registro_de_datos:
            print("Error: No hay pacientes registrados. Registre al menos un paciente antes de generar reportes.")
            continue
        print("\nSeccion de Consultas y Reportes:")
        print("1.- Reporte de Citas por Período\n"
              "2.- Reporte de Pacientes Registrados\n"
              "3.- Listado completo de pacientes\n"
              "4.- Buscar Paciente por Clave\n"
              "5.- Buscar Pacientes por Apellidos y Nombres\n")
        try:
            opcion_reporte = int(input("Ingrese la opcion de consulta o reporte que desea realizar: "))
        except ValueError:
            print("Debera de ser un valor numerico y dentro del rango de opciones")
            continue

        if opcion_reporte == 1:
            generar_reporte_citas_por_periodo()

        elif opcion_reporte == 2:
            generar_reporte_pacientes()

        elif opcion_reporte == 3:
            listado_completo_pacientes()

        elif opcion_reporte == 4:
            buscar_paciente_por_clave()

        elif opcion_reporte == 5:
                buscar_pacientes_por_apellidos_y_nombres()

        else:
            print("Opcion no valida")

    elif opcion_menu == 5:
        if confirmar_salida():
            print("Saliendo del sistema...")
            break

    else:
        print("Opcion no valida. Por favor, seleccione una opcion del 1 al 5.")
```
